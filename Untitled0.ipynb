```python
import matplotlib.pyplot as plt
```

```python
# Maʼlumotlar toʻplamini yuklang (“cars.csv” xuddi shu katalogda mavjud boʻlsa)
df = pd.read_csv('/content/drive/MyDrive/content /car_dataset.csv')
# DataFrame-da "Model" va "Yurgan yo'li" ustunlari mavjudligini tekshiring

if 'Model' in df.columns and 'Yurgan yo\'li' in df.columns:
    # "Yurgan yo'li" raqamli ustun ekanligiga ishonch hosil qiling
    df['Yurgan yo\'li'] = pd.to_numeric(df['Yurgan yo\'li'], errors='coerce')
    # "Yurgan yo'li" ustunida qiymatlari etishmayotgan qatorlarni tashlang
    df.dropna(subset=['Yurgan yo\'li'], inplace=True)

    # DataFrame-ni "Yurgan yo'li" boʻyicha tartiblang
    df.sort_values('Yurgan yo\'li', inplace=True)
    models = df['Model']
    kilometers_driven = df['Yurgan yo\'li']

    # Diagramma yaratish
    fig, ax = plt.subplots(figsize=(10, 8))
    y_pos = np.arange(len(models))
    bars = ax.barh(y_pos, kilometers_driven, align='center', color='skyblue', edgecolor='black')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(models)
    ax.invert_yaxis()  # Y tengligini o'zgartirish (yuqoridan pastga qarab)
    ax.set_xlabel('Yurgan yo`li')
    ax.set_title('Mashina modelining yurgan yo`li diagrammasi')

    # Har bir bar uchun qiymatlarni text shaklida qo'shish
    for bar, km in zip(bars, kilometers_driven):
        ax.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{km}', va='center', ha='left', fontsize=10, color='black')

    plt.show()
else:
    print("Ma'lumotlar toʻplamida \"Model\" yoki \"Yurgan yo'li\" ustuni mavjud emas.")
```
